```python
# locating the rows with outliers and it has signifance

column_data = df['Total Amount']

mean = np.mean(column_data)
std_dev = np.std(column_data)
distances = np.abs(column_data - mean) / std_dev
outliers = distances > 2

print(df[outliers])
```
